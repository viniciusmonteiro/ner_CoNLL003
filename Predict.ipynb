```python
from simpletransformers.ner import NERModel,NERArgs
from os import path


# NER model
chosen_pretrained = 'dslim/bert-base-NER'
model = NERModel('bert', chosen_pretrained, use_cuda=False)

sentence = "This sentence was written in Italy by Vinicius Lira, but it could be written in Switzerland by someone else"
prediction, model_output = model.predict([sentence])
```
